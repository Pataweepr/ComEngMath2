```
import numpy as np 
from scipy import signal,fftpack
import matplotlib.pyplot as plt
import soundfile as sf
import IPython.display as ipd
%matplotlib inline
import librosa
import os
```

```
path = os.path.join('..','collect_song','11. TSU・BA・SA.flac')
data, sr = sf.read(path)

print('sampling_rate: {} Hz.'.format(sr))

t_start = 0.5
t_end = 3.25
sel_start = int(t_start * sr)
sel_end = int(t_end * sr)
plt.figure(figsize=(20, 5))
plt.plot(np.arange(sel_start,sel_end)/sr,data[sel_start:sel_end,:])
plt.show()
```

```
ipd.Audio(data[sel_start:sel_end,0], rate=sr)
```

```
samp_point = 100
re_samp_point = 50
xnew_test = np.linspace(0, 10, samp_point, endpoint=False)
y_test = np.cos(-xnew_test**2/6.0)
f = signal.resample(y_test, re_samp_point)
x = np.linspace(0, 10, re_samp_point, endpoint=False)
plt.axhline(y_test[0], color='red', lw=2)
plt.plot(xnew_test, y_test, 'go-', x, f, '.-')
plt.show()
```

```
t_st = 0 
t_end = 0.05
f_01 = 1200
f_02 = 1300

f_samp_01 = 5000
f_samp_02 = 2500

ts01 = np.linspace(t_st, t_end , int((t_end-t_st)*f_samp_01), endpoint=False)
x01_ts01 = np.sin(2*np.pi*f_01*ts01)
x02_ts01 = np.sin(2*np.pi*f_02*ts01)


plt.figure(figsize=(20, 5))
plt.plot(ts01, x01_ts01, 'go-', ts01, x02_ts01, 'r.-')
plt.show()

ts02 = np.linspace(t_st, t_end , int((t_end-t_st)*f_samp_02), endpoint=False)
x01_ts02 = np.sin(2*np.pi*f_01*ts02)
x02_ts02 = np.sin(2*np.pi*f_02*ts02)


plt.figure(figsize=(20, 5))
plt.plot(ts02, x01_ts02, 'go-', ts02, x02_ts02, 'r.-')
plt.show()
```
