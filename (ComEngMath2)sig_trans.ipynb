```
import numpy as np 
from scipy import signal
from scipy import fftpack
import matplotlib.pyplot as plt

import IPython.display as ipd
%matplotlib inline
import librosa
import os
```

# Time scaling

```
t = np.linspace(0, 1, 500)
plt.plot(t, signal.sawtooth(2 * np.pi * 5 * t))


scalring = 3
t = np.linspace(0, 1, 500)
plt.plot(t, signal.sawtooth(2 * np.pi * 5 * t/scalring))
plt.show()
```

# Time Shifting

```
plt.plot(t, signal.square(2 * np.pi * 5 * t))

shift = 0.05
plt.plot(t, signal.square(2 * np.pi * 5 * (t-shift)))
plt.ylim(-2, 2)
plt.show()
```

# Relction signal

```
# reflec
t = np.linspace(0, 1, 500)
plt.plot(t, signal.sawtooth(2 * np.pi * 5 * t))


t = np.linspace(0, 1, 500)
plt.plot(t, signal.sawtooth(2 * np.pi * 5 * (1/5-t)))
plt.show()
```
